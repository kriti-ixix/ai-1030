```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

# Harris Corner Detection

```python
# Convert images to float
gray = np.float32(gray_flat_chess)

# Harris Corner
result = cv2.cornerHarris(src = gray, blockSize=2, ksize=3, k=0.04)

# Dilating the corners
result = cv2.dilate(result, None)

# Thresholding
flat_chess[result > 0.01 * result.max()] = [255, 0, 0]
plt.imshow(flat_chess)
```

```python
# Convert images to float
gray = np.float32(gray_real_chess)

# Harris Corner
result = cv2.cornerHarris(src = gray, blockSize=2, ksize=3, k=0.04)

# Dilating the corners
result = cv2.dilate(result, None)

# Thresholding
real_chess[result > 0.01 * result.max()] = [255, 0, 0]
plt.imshow(real_chess)
```

# Shi-Tomasi Corner Detection

```python
flat_chess = cv2.imread('flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 5, 0.01, 10)
corners = np.int0(corners)

for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, 255, -1)
    
plt.imshow(flat_chess)
```
