```python
import cv2
import matplotlib.pyplot as plt
```

# Canny Edge Detection

```python
squirtle = cv2.imread('squirtle.webp')
plt.imshow(squirtle)
```

```python
edges = cv2.Canny(image=squirtle, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
doggy = cv2.imread('sammy_face.jpg')
plt.imshow(doggy)
```

```python
edges = cv2.Canny(image=doggy, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=doggy, threshold1=200, threshold2=255)
plt.imshow(edges)
```

# Blurred Images

```python
blurred = cv2.GaussianBlur(doggy, (5, 5), cv2.BORDER_DEFAULT)
plt.imshow(blurred)
```

```python
edges = cv2.Canny(image=blurred, threshold1=127, threshold2=127)
plt.imshow(edges)
```

# Erosion and Dilation

```python
plt.imshow(doggy)
```

```python
plt.imshow(edges, cmap='gray')
```

```python
dilated_image = cv2.dilate(edges, (3, 3), iterations=1)
plt.imshow(dilated_image, cmap='gray')
```

```python
eroded_image = cv2.erode(edges, (3, 3), iterations=1)
plt.imshow(eroded_image, cmap='gray')
```
