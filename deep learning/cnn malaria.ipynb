# Checking for dimensions

```python
dim_1 = []
dim_2 = []
```

```python
for image in os.listdir(train_path + "/uninfected"):
    image_path = train_path + "/uninfected/" + image
    img = im.imread(image_path)
    d1, d2, c = img.shape
    dim_1.append(d1)
    dim_2.append(d2)
```

```python
for image in os.listdir(train_path + "/parasitized"):
    image_path = train_path + "/parasitized/" + image
    # print(image_path)
    img = im.imread(image_path)
    d1, d2, c = img.shape
    dim_1.append(d1)
    dim_2.append(d2)
```

```python
plt.scatter(dim_1, dim_2)
```

# Data Augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_generator=ImageDataGenerator(rotation_range=20,width_shift_range=0.1,rescale=1/255,
                  height_shift_range=0.1,shear_range=0.1,zoom_range=0.1,
                  horizontal_flip=True,fill_mode="nearest")
```

```python
plt.imshow(para_arr)
```

```python
plt.imshow(img_generator.random_transform(para_arr))
```

# Reading the Images

```python
os.listdir(train_path)
```

```python
train_img = []
train_label = []

test_img = []
test_label = []

for folder in os.listdir(train_path):
    for image in os.listdir(train_path + "/" + folder):
        img = im.imread(train_path + "/" + folder + "/" + image)
        train_img.append(img)
        train_label.append(folder)
        
for folder in os.listdir(test_path):
    for image in os.listdir(test_path + "/" + folder):
        img = im.imread(test_path + "/" + folder + "/" + image)
        test_img.append(img)
        test_label.append(folder)
```
